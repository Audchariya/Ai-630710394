เราจะได้ข้อมูลที่สมบูรณ์ในระดับนึง พร้อมที่จะทำขั้นตอนการ Cleaning ต่อไป

### วิธีที่ 2 เติมค่าด้วยค่าเฉลี่ย

### 1. การลบแถวที่มีข้อมูลขาดหายมากเกินไป
- df_std.isnull().sum(axis=1) คำนวณจำนวนค่าที่หาย (NaN) ในแต่ละแถว (axis=1)
- <= 3 ตรวจสอบว่าแถวนั้นมีค่าที่หายไม่เกิน 3 คอลัมน์
- สร้าง DataFrame ใหม่ df_clean ที่เก็บเฉพาะแถวที่ตรงตามเงื่อนไข
- .copy() เพื่อป้องกันการแก้ไขข้อมูลโดยไม่ตั้งใจใน DataFrame เดิม

### 2. เติมค่าที่หายในคอลัมน์ตัวเลขด้วยค่าเฉลี่ย
- select_dtypes(include=['float64', 'int64']) เลือกคอลัมน์ที่เป็นประเภทตัวเลข
- .columns ได้รายชื่อคอลัมน์เหล่านั้น
- สำหรับแต่ละคอลัมน์ตัวเลข: เติมค่าที่หาย (NaN) ด้วยค่าเฉลี่ยของคอลัมน์นั้น (mean())

### 3. เติมค่าที่หายในคอลัมน์หมวดหมู่ด้วยค่าที่พบบ่อยที่สุด
- select_dtypes(include=['object']) เลือกคอลัมน์ที่เป็นประเภทข้อความ/หมวดหมู่
- .columns ได้รายชื่อคอลัมน์เหล่านั้น
- สำหรับแต่ละคอลัมน์หมวดหมู่: เติมค่าที่หาย (NaN) ด้วยค่าที่ปรากฏบ่อยที่สุด (mode()[0])


```python
# 1. ลบแถวที่หายมากกว่า 3 คอลัมน์
df_clean = df_std[df_std.isnull().sum(axis=1) <= 3].copy()

# 2. เติมค่าด้วยค่าเฉลี่ยสำหรับตัวเลข
num_cols = df_clean.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mean())

# 3. เติมค่าด้วยค่าที่พบบ่อยที่สุดสำหรับหมวดหมู่
cat_cols = df_clean.select_dtypes(include=['object']).columns
for col in cat_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mode()[0])
```

```python
# ปรับทศนิยมเป็น 2 ตำแหน่ง
df_clean = df_clean.round(2)

# ลบทศนิยม (ปัดเป็นจำนวนเต็ม)
df_clean["อายุ"] = df_clean["อายุ"].round(0).astype(int)
df_clean["ชั่วโมงเรียน_ต่อสัปดาห์"] = df_clean["ชั่วโมงเรียน_ต่อสัปดาห์"].round(0).astype(int)
```

##2. การทำให้ข้อมูลเป็นมาตรฐาน
เหตุผลที่ต้องทำขั้นตอนนี้

คือเพื่อให้ข้อมูลที่มีความหมายเหมือนกันถูกเขียนในรูปแบบเดียวกันทั้งหมด เพื่อป้องกันปัญหาที่เกิดจากความไม่สม่ำเสมอ และ ขั้นตอนนี้ทำให้ข้อมูล สะอาด + สม่ำเสมอ + พร้อมใช้งาน โดยไม่เกิดความเข้าใจผิดจากความแตกต่างในรูปแบบการเขียน

⬇️โค้ดนี้มีไว้ทำให้ชื่อคณะเขียนเหมือนกันหมด เช่น "วิทย์", "Science" หรือ "คณะวิทยาศาสตร์" ก็เปลี่ยนให้เป็น "วิทยาศาสตร์" รวมถึงเพศก็เช่นกัน จากนั้นใช้ .apply() ให้ฟังก์ชันนี้ทำงานกับทุกแถวในคอลัมน์ คณะ แล้วเก็บผลไว้ในคอลัมน์ใหม่ชื่อ คณะ_มาตรฐาน เพื่อให้ข้อมูลคณะเป็นรูปแบบเดียวกันและนำไปวิเคราะห์ต่อได้ง่าย

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์' , 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวะ', 'Engineering', 'วิศวกรรม' , 'วิศวฯ',]:
        return 'วิศวกรรมศาสตร์'
    else:
        return name

# ใช้งาน
df_std['คณะ_มาตรฐาน'] = df_std['คณะ'].apply(standardize_faculty)
```

```python
def standardize_gender(gender):
    if pd.isna(gender):
        return gender
    if gender in ['ชาย', 'Male', 'M', 'ช']:
        return 'ชาย'
    elif gender in ['หญิง', 'Female', 'F', 'ญ']:
        return 'หญิง'
    else:
        return gender

df_std['เพศ_มาตรฐาน'] = df_std['เพศ'].apply(standardize_gender)
```

##3. การจัดการค่าผิดปกติ
ต้องจัดการค่าผิดปกติ (Outliers) เพราะช่วยให้ข้อมูลสะท้อนความจริงมากขึ้น ค่าหลุดสูงหรือต่ำเกินไปจะทำให้ค่าเฉลี่ยและการกระจายข้อมูลเพี้ยน ส่งผลให้การวิเคราะห์และโมเดลทำนายผิด ความสัมพันธ์ระหว่างตัวแปรคลาดเคลื่อน อีกทั้งหลายกรณีเกิดจากการกรอกข้อมูลผิดหรือระบบบันทึกผิด การจัดการจึงช่วยเพิ่มความน่าเชื่อถือ ลดผลกระทบต่อโมเดล และทำให้การแสดงผลกราฟหรือสถิติมีสเกลที่อ่านง่าย เห็นรูปแบบข้อมูลปกติชัดเจนขึ้น

```python
print("ข้อมูลต้นฉบับ (สรุป)")
print(df_std["อายุ"].describe())

# คำนวณ IQR
Q1 = df_std["อายุ"].quantile(0.25)
Q3 = df_std["อายุ"].quantile(0.75)
IQR = Q3 - Q1

# ขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"\nขอบเขตล่าง: {lower_bound}")
print(f"ขอบเขตบน: {upper_bound}")

# กรองค่าผิดปกติออกไป
df_no_outliers = df_std[(df_std["อายุ"] >= lower_bound) & (df_std["อายุ"] <= upper_bound)]

print("\nข้อมูลหลังกรองอายุ (สรุป)")
print(df_no_outliers["อายุ"].describe())

df_std = df_no_outliers
```

```python
# แก้ไข GPA ผิดปกติ
df_std.loc[df_std['GPA'] < 0, 'GPA'] = np.nan
df_std.loc[df_std['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df_std['GPA'].mean()

df_std['GPA'].fillna(mean_gpa, inplace=True)
```

#Feature Engineering - การสร้างคุณลักษณะใหม่

##1. One-Hot Encoding แปลงข้อมูลประเภทเป็นตัวเลข bold text
1) มาตรฐานค่า “เพศ” เพราะข้อมูลเพศมีหลายแบบที่ความหมายเหมือนกัน เช่น ชาย, Male, M, ช หมายถึง “ชาย” เหมือนกัน เราเลยใช้ gender_map กับ .replace() เพื่อให้เป็นคำเดียวกัน จะได้ไม่งงตอนวิเคราะห์

2) มาตรฐานค่า “คณะ” ชื่อคณะก็มีหลายแบบ เช่น วิทยาศาสตร์, วิทย์, Science, คณะวิทยาศาสตร์ หมายถึงคณะเดียวกัน ก็ใช้ faculty_map รวมให้เหลือรูปแบบเดียว เพื่อป้องกันคอลัมน์ซ้ำ ๆ ตอนทำต่อ

3) One-Hot Encoding พอค่ามาตรฐานแล้ว เราใช้ pd.get_dummies() แปลงข้อมูลจากข้อความเป็นตัวเลข 0/1 เช่น คอลัมน์ เพศ จะกลายเป็น เพศ_ชาย กับ เพศ_หญิง ถ้าเพศตรงกับชื่อคอลัมน์ก็จะเป็น 1 ถ้าไม่ตรงก็เป็น 0 ทำให้เอาไปใช้กับโมเดลหรือคำนวณได้ง่าย

```python
# 1) มาตรฐานค่าของ "เพศ"
gender_map = {
    "ชาย": "ชาย", "Male": "ชาย", "M": "ชาย", "ช": "ชาย",
    "หญิง": "หญิง", "Female": "หญิง", "F": "หญิง", "ญ": "หญิง"
}
df_std["เพศ"] = df_std["เพศ"].replace(gender_map)

# -------------------------
# 2) มาตรฐานค่าของ "คณะ"
faculty_map = {
    "วิทยาศาสตร์": "วิทยาศาสตร์", "วิทย์": "วิทยาศาสตร์", "Science": "วิทยาศาสตร์", "คณะวิทยาศาสตร์": "วิทยาศาสตร์",
    "วิศวะ": "วิศวกรรมศาสตร์", "Engineering": "วิศวกรรมศาสตร์", "วิศวกรรม": "วิศวกรรมศาสตร์", "วิศวฯ": "วิศวกรรมศาสตร์", "วิศวกรรมศาสตร์": "วิศวกรรมศาสตร์",
    "ศิลปศาสตร์": "ศิลปศาสตร์", "Arts": "ศิลปศาสตร์", "ศิลป์": "ศิลปศาสตร์", "Liberal Arts": "ศิลปศาสตร์", "Sci": "วิทยาศาสตร์"
}
df_std["คณะ"] = df_std["คณะ"].replace(faculty_map)

# -------------------------
# 3) One-Hot Encoding
df_encoded = pd.get_dummies(df_std, columns=["เพศ", "คณะ"], prefix=["เพศ", "คณะ"])
```

## 2. การสร้าง Feature จากวันเวลา
- ส่วนที่ 1 เราจะแปลงคอลัมน์วันที่ให้เป็นรูปแบบ datetime
  - pd.to_datetime() แปลงคอลัมน์ "วันที่ลงทะเบียน" ให้เป็นรูปแบบ datetime ของ    pandas

  - พารามิเตอร์ errors="coerce" หมายความว่าถ้าวันที่แปลงไม่ได้ (เช่นรูปแบบไม่ถูกต้อง) จะถูกเปลี่ยนเป็นค่า NaT (Not a Time) แทนที่จะเกิด error

  - ผลลัพธ์จะถูกบันทึกกลับไปที่คอลัมน์เดิม "วันที่ลงทะเบียน"

- ส่วนที่ 2: สร้างคอลัมน์วันในสัปดาห์ภาษาไทย
  - thai_days เป็นลิสต์ที่เก็บชื่อวันในสัปดาห์ภาษาไทย
  - df_std["วันที่ลงทะเบียน"].dt.dayofweek ได้เลข index ของวันในสัปดาห์
  - enumerate(thai_days) สร้าง tuple ของ (index, ชื่อวัน) เช่น (0, 'จันทร์'), (1, 'อังคาร') เป็นต้น
  - dict(enumerate(thai_days)) แปลงเป็น dictionary สำหรับ mapping
  - .map() ใช้ dictionary นี้เพื่อแปลงเลข index เป็นชื่อวันภาษาไทย
  - ผลลัพธ์ถูกบันทึกในคอลัมน์ใหม่ชื่อ "วันในสัปดาห์_ลงทะเบียน"

```python
df_std["วันที่ลงทะเบียน"] = pd.to_datetime(df_std["วันที่ลงทะเบียน"], errors="coerce")


thai_days = ['จันทร์','อังคาร','พุธ','พฤหัสบดี','ศุกร์','เสาร์','อาทิตย์']
df_std["วันในสัปดาห์_ลงทะเบียน"] = df_std["วันที่ลงทะเบียน"].dt.dayofweek.map(dict(enumerate(thai_days)))
```

#3. การรวม Features

โค้ดนี้ทำเพราะต้องการสร้างฟีเจอร์ BMI และจัดกลุ่มสุขภาพจากข้อมูลส่วนสูงและน้ำหนัก เพื่อใช้วิเคราะห์หรือเปรียบเทียบสุขภาพของนักศึกษาได้ง่ายขึ้น

สิ่งที่ทำคือ

1.แปลงส่วนสูงจากเซนติเมตรเป็นเมตร และกันค่าศูนย์เพื่อเลี่ยงหารด้วยศูนย์

2.คำนวณค่า BMI ตามสูตร น้ำหนัก ÷ (ส่วนสูง²) และปัดเป็นทศนิยม 2 ตำแหน่ง

3.แบ่งค่า BMI ออกเป็นช่วง พร้อมใส่ป้ายกำกับ เช่น ผอม, ปกติ, อ้วน, อ้วนมาก

```python
# ป้องกันหารด้วยศูนย์/ค่าว่าง
h_w = (df_std["ส่วนสูง_cm"] / 100).replace(0, np.nan)

df_std["BMI"] = df_std["น้ำหนัก_kg"] / (h_w ** 2)
df_std["BMI"] = df_std["BMI"].round(2)

df_std["สุขภาพ"] = pd.cut(
    df_std["BMI"],
    bins=[0, 18.5, 25, 30, np.inf],
    labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'],
    right=True, include_lowest=True
)
```
